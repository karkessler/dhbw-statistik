# Häufigkeiten und Diagramme

Dieses Notebook berechnet aus absoluten Häufigkeiten
die relativen Häufigkeiten (%) und die Summenhäufigkeit (%).

Diagramme:
- Balkendiagramm: relative Häufigkeit
- Liniendiagramm: Summenhäufigkeit


```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({
    "figure.dpi": 120,
    "axes.spines.top": False,
    "axes.spines.right": False
})
```

## Daten eingeben
Noten: 1–5
Absolute Häufigkeiten: 19, 57, 132, 137, 204


```python
noten = [1, 2, 3, 4, 5]
absolute = [19, 57, 132, 137, 204]

df = pd.DataFrame({
    "Note": noten,
    "Absolute Häufigkeit": absolute
})

n = df["Absolute Häufigkeit"].sum()
df["Relative Häufigkeit (%)"] = df["Absolute Häufigkeit"] / n * 100
df["Summenhäufigkeit (%)"] = df["Relative Häufigkeit (%)"].cumsum()

df.round(2)
```

## Balkendiagramm – Relative Häufigkeit


```python
plt.figure(figsize=(6, 4))
plt.bar(df["Note"], df["Relative Häufigkeit (%)"])
plt.xlabel("Note")
plt.ylabel("Anteil in %")
plt.title("Relative Häufigkeit (%)")
plt.show()
```

## Liniendiagramm – Summenhäufigkeit


```python
plt.figure(figsize=(6, 4))
plt.plot(df["Note"], df["Summenhäufigkeit (%)"], marker="o")
plt.xlabel("Note")
plt.ylabel("Anteil kumuliert in %")
plt.ylim(0, 105)
plt.title("Summenhäufigkeit (%)")
plt.show()
```

## Interpretation


```python
anteil_bis_3 = df.loc[df["Note"] == 3, "Summenhäufigkeit (%)"].iloc[0]
print(f"{anteil_bis_3:.2f} % der Studierenden haben Note 3 oder besser.")
```
